Week 1 Assignment: Transfer Learning & Deep Neural Network

```python
"""
Part 1: DenseNet-121 Feature Extraction on CIFAR-10
Part 2: Deep Neural Network (No CNN) on MNIST

"""

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
import torchvision
import torchvision.transforms as transforms
from torchvision import datasets, models
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")


# PART 1: DenseNet-121 Transfer Learning on CIFAR-10

print("\n" + "="*60)
print("PART 1: DenseNet-121 Feature Extraction on CIFAR-10")
print("="*60)

# Data Preparation for CIFAR-10

# ImageNet normalization values
cifar_transform_train = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(32, padding=4),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                        std=[0.229, 0.224, 0.225])
])

cifar_transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                        std=[0.229, 0.224, 0.225])
])

# Load CIFAR-10 dataset
print("\nLoading CIFAR-10 dataset...")
cifar_train_full = datasets.CIFAR10(root='./data', train=True,
                                     download=True, transform=cifar_transform_train)
cifar_test = datasets.CIFAR10(root='./data', train=False,
                               download=True, transform=cifar_transform_test)

# Split training set into train and validation (80-20 split)
train_size = int(0.8 * len(cifar_train_full))
val_size = len(cifar_train_full) - train_size
cifar_train, cifar_val = random_split(cifar_train_full, [train_size, val_size])

# Create data loaders
batch_size = 128
cifar_train_loader = DataLoader(cifar_train, batch_size=batch_size,
                                shuffle=True, num_workers=2)
cifar_val_loader = DataLoader(cifar_val, batch_size=batch_size,
                              shuffle=False, num_workers=2)
cifar_test_loader = DataLoader(cifar_test, batch_size=batch_size,
                               shuffle=False, num_workers=2)

print(f"Train samples: {len(cifar_train)}")
print(f"Validation samples: {len(cifar_val)}")
print(f"Test samples: {len(cifar_test)}")

# Build DenseNet-121 Model
# -----------------------------
class DenseNetFeatureExtractor(nn.Module):
    def __init__(self, num_classes=10):
        super(DenseNetFeatureExtractor, self).__init__()
        # Load pretrained DenseNet-121
        self.densenet = models.densenet121(weights='DEFAULT')

        # Freeze all convolutional layers
        for param in self.densenet.parameters():
            param.requires_grad = False

        # Get number of features from the classifier
        num_features = self.densenet.classifier.in_features

        # Replace classifier with custom head
        # Adding BatchNorm for better stability
        self.densenet.classifier = nn.Sequential(
            nn.BatchNorm1d(num_features),
            nn.Dropout(0.3),
            nn.Linear(num_features, 512),
            nn.ReLU(),
            nn.BatchNorm1d(512),
            nn.Dropout(0.3),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        return self.densenet(x)

print("\nBuilding DenseNet-121 model...")
densenet_model = DenseNetFeatureExtractor(num_classes=10).to(device)
print("Model architecture:")
print(f"Total parameters: {sum(p.numel() for p in densenet_model.parameters()):,}")
print(f"Trainable parameters: {sum(p.numel() for p in densenet_model.parameters() if p.requires_grad):,}")

# Training Setup
# ------------------
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(densenet_model.parameters(), lr=0.001)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min',
                                                  factor=0.5, patience=3)

# Early stopping setup
class EarlyStopping:
    def __init__(self, patience=7, min_delta=0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = None
        self.early_stop = False

    def __call__(self, val_loss):
        if self.best_loss is None:
            self.best_loss = val_loss
        elif val_loss > self.best_loss - self.min_delta:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_loss = val_loss
            self.counter = 0

early_stopping = EarlyStopping(patience=5)

# Training Function
# ---------------------
def train_epoch(model, loader, criterion, optimizer):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    pbar = tqdm(loader, desc='Training')
    for inputs, labels in pbar:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

        pbar.set_postfix({'loss': running_loss/len(loader),
                         'acc': 100.*correct/total})

    return running_loss/len(loader), 100.*correct/total

def validate(model, loader, criterion):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    return running_loss/len(loader), 100.*correct/total

# Train DenseNet Model
# ------------------------
print("\nTraining DenseNet-121...")
num_epochs = 20
densenet_train_losses = []
densenet_val_losses = []
densenet_train_accs = []
densenet_val_accs = []

for epoch in range(num_epochs):
    print(f"\nEpoch {epoch+1}/{num_epochs}")

    train_loss, train_acc = train_epoch(densenet_model, cifar_train_loader,
                                        criterion, optimizer)
    val_loss, val_acc = validate(densenet_model, cifar_val_loader, criterion)

    densenet_train_losses.append(train_loss)
    densenet_val_losses.append(val_loss)
    densenet_train_accs.append(train_acc)
    densenet_val_accs.append(val_acc)

    print(f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%")
    print(f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")

    scheduler.step(val_loss)
    early_stopping(val_loss)

    if early_stopping.early_stop:
        print("\nEarly stopping triggered!")
        break

# Evaluate DenseNet on Test Set
# ----------------------------------
print("\nEvaluating DenseNet-121 on test set...")
test_loss, test_acc = validate(densenet_model, cifar_test_loader, criterion)
print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_acc:.2f}%")

# Get predictions for confusion matrix
densenet_model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for inputs, labels in cifar_test_loader:
        inputs = inputs.to(device)
        outputs = densenet_model(inputs)
        _, predicted = outputs.max(1)
        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.numpy())

# Plot confusion matrix
cifar_classes = ['airplane', 'automobile', 'bird', 'cat', 'deer',
                 'dog', 'frog', 'horse', 'ship', 'truck']

cm_densenet = confusion_matrix(all_labels, all_preds)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(densenet_train_losses, label='Train Loss')
plt.plot(densenet_val_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('DenseNet-121: Training & Validation Loss')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(densenet_train_accs, label='Train Accuracy')
plt.plot(densenet_val_accs, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('DenseNet-121: Training & Validation Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig('densenet_training_curves.png', dpi=300, bbox_inches='tight')
plt.show()

# Plot confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cm_densenet, annot=True, fmt='d', cmap='Blues',
            xticklabels=cifar_classes, yticklabels=cifar_classes)
plt.title('DenseNet-121 Confusion Matrix on CIFAR-10')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.savefig('densenet_confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "="*60)
print("PART 1 COMPLETE: DenseNet-121 Results")
print("="*60)
print(f"Final Test Accuracy: {test_acc:.2f}%")


# PART 2: Deep Neural Network (No CNN) on MNIST

print("\n" + "="*60)
print("PART 2: Deep Neural Network on MNIST")
print("="*60)

# Data Preparation for MNIST

# Data augmentation for MNIST
mnist_transform_train = transforms.Compose([
    transforms.RandomRotation(10),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

mnist_transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

print("\nLoading MNIST dataset...")
mnist_train_full = datasets.MNIST(root='./data', train=True,
                                  download=True, transform=mnist_transform_train)
mnist_test = datasets.MNIST(root='./data', train=False,
                            download=True, transform=mnist_transform_test)

# Split training set into train and validation (80-20 split)
train_size = int(0.8 * len(mnist_train_full))
val_size = len(mnist_train_full) - train_size
mnist_train, mnist_val = random_split(mnist_train_full, [train_size, val_size])

# Create data loaders
batch_size = 128
mnist_train_loader = DataLoader(mnist_train, batch_size=batch_size,
                                shuffle=True, num_workers=2)
mnist_val_loader = DataLoader(mnist_val, batch_size=batch_size,
                              shuffle=False, num_workers=2)
mnist_test_loader = DataLoader(mnist_test, batch_size=batch_size,
                               shuffle=False, num_workers=2)

print(f"Train samples: {len(mnist_train)}")
print(f"Validation samples: {len(mnist_val)}")
print(f"Test samples: {len(mnist_test)}")

# Build Deep Neural Network

class DeepNeuralNetwork(nn.Module):
    def __init__(self, input_size=784, hidden_sizes=[512, 256, 128], num_classes=10):
        super(DeepNeuralNetwork, self).__init__()

        layers = []
        prev_size = input_size

        for hidden_size in hidden_sizes:
            layers.extend([
                nn.Linear(prev_size, hidden_size),
                nn.BatchNorm1d(hidden_size),
                nn.ReLU(),
                nn.Dropout(0.3)
            ])
            prev_size = hidden_size

        # Output layer
        layers.append(nn.Linear(prev_size, num_classes))

        self.network = nn.Sequential(*layers)

    def forward(self, x):
        # Flatten the input (batch_size, 1, 28, 28) -> (batch_size, 784)
        x = x.view(x.size(0), -1)
        return self.network(x)

print("\nBuilding Deep Neural Network...")
dnn_model = DeepNeuralNetwork().to(device)
print("Model architecture:")
print(dnn_model)
print(f"Total parameters: {sum(p.numel() for p in dnn_model.parameters()):,}")

# Training Setup
# ------------------
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(dnn_model.parameters(), lr=0.001)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min',
                                                  factor=0.5, patience=3)

early_stopping = EarlyStopping(patience=7)

# Train DNN Model
# -------------------
print("\nTraining Deep Neural Network...")
num_epochs = 30
dnn_train_losses = []
dnn_val_losses = []
dnn_train_accs = []
dnn_val_accs = []

for epoch in range(num_epochs):
    print(f"\nEpoch {epoch+1}/{num_epochs}")

    train_loss, train_acc = train_epoch(dnn_model, mnist_train_loader,
                                        criterion, optimizer)
    val_loss, val_acc = validate(dnn_model, mnist_val_loader, criterion)

    dnn_train_losses.append(train_loss)
    dnn_val_losses.append(val_loss)
    dnn_train_accs.append(train_acc)
    dnn_val_accs.append(val_acc)

    print(f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%")
    print(f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")

    scheduler.step(val_loss)
    early_stopping(val_loss)

    if early_stopping.early_stop:
        print("\nEarly stopping triggered!")
        break

# Evaluate DNN on Test Set
# -----------------------------
print("\nEvaluating DNN on test set...")
test_loss, test_acc = validate(dnn_model, mnist_test_loader, criterion)
print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_acc:.2f}%")

# Get predictions for confusion matrix
dnn_model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for inputs, labels in mnist_test_loader:
        inputs = inputs.to(device)
        outputs = dnn_model(inputs)
        _, predicted = outputs.max(1)
        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.numpy())

cm_dnn = confusion_matrix(all_labels, all_preds)

# Plot training curves
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(dnn_train_losses, label='Train Loss')
plt.plot(dnn_val_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('DNN: Training & Validation Loss')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(dnn_train_accs, label='Train Accuracy')
plt.plot(dnn_val_accs, label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('DNN: Training & Validation Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig('dnn_training_curves.png', dpi=300, bbox_inches='tight')
plt.show()

# Plot confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cm_dnn, annot=True, fmt='d', cmap='Greens',
            xticklabels=range(10), yticklabels=range(10))
plt.title('Deep Neural Network Confusion Matrix on MNIST')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.savefig('dnn_confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "="*60)
print("PART 2 COMPLETE: Deep Neural Network Results")
print("="*60)
print(f"Final Test Accuracy: {test_acc:.2f}%")
```


# Observations


For Part 1 (DenseNet-121 on CIFAR-10):
- Transfer learning significantly improved accuracy compared to training from scratch
- Freezing convolutional layers reduced training time dramatically
- BatchNorm in the classifier head stabilized training
- Data augmentation (flips, crops) helped prevent overfitting
- Learning rate scheduling improved convergence
- Early stopping prevented overfitting when validation loss plateaued

For Part 2 (DNN on MNIST):
- Deep architecture (3 hidden layers) captured complex patterns
- BatchNorm significantly improved training stability
- Dropout (0.3) effectively prevented overfitting
- Data augmentation (rotation, translation) improved generalization
- Adam optimizer with learning rate scheduling worked well
- Early stopping saved training time

------
- CIFAR-10 is more difficult than MNIST (lower accuracy expected)
- Without augmentation, both models overfit quickly
- Too much dropout (>0.5) hurt training speed
- Very deep networks (>4 layers) didn't help for MNIST
